```python
import matplotlib as mpl # in python
import matplotlib.cm as cmx
cNorm  = mpl.colors.Normalize(vmin=0,vmax=len(june_files))
Map = cmx.ScalarMappable(norm=cNorm, cmap='viridis')
Map._A = []
```

```python
lfiles = june_files

cNorm  = mpl.colors.Normalize(vmin=0,vmax=len(lfiles))
Map = cmx.ScalarMappable(norm=cNorm, cmap='viridis')
Map._A = []

figd, axd = plt.subplots()
figdAr, axdAr = plt.subplots()
figr, axr = plt.subplots()
for i,jf in enumerate(lfiles):
    df = rr.realign(rr.IO_orb(rr.read(jf)))
    axd.plot(df['abundance_CO2'],df['alt'],color=Map.to_rgba(i))
    axdAr.plot(df['abundance_Ar'],df['alt'],color=Map.to_rgba(i))
    axr.plot(df['abundance_N2']/df['abundance_Ar'],df['alt'],color=Map.to_rgba(i))
axd.set_xscale('log')
axdAr.set_xscale('log')
axr.set_xscale('log')

#axr.set_ylim(155,200)
# axr.set_xlim(2.,20)
plt.show()
```
