```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("../../maven/data/sci/ngi/l2/2018/09/mvn_ngi_l2_ion-abund-28876_20180909T180702_v08_r01.csv")
```

```python
co2p = df[df["ion_mass"] == 44]
```

```python
year = 2018
month = 6
files = glob.glob(f"../../maven/data/sci/ngi/l2/{year}/{str(month).zfill(2)}/mvn_ngi_l2_ion-abund-*.csv")
```

```python
plt.figure(figsize=(8,8))
for f in files:
    df = pd.read_csv(f)
    co2p = df[df["ion_mass"] == 44]
    plt.plot(co2p["abundance"], co2p["alt"], lw=0.5, alpha=0.5)
plt.xscale("log")
plt.xlabel("Abudndance [cm^-3]")
plt.ylabel("")
plt.ylim(110, 400)

plt.savefig(f"../../Desktop/CO2p_{year}{str(month).zfill(2)}.png", dpi=300)

#plt.show()
```
